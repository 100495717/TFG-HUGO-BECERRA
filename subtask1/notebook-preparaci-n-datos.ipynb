```python
import os
import csv

# Definimos la ruta base donde están las carpetas train, dev y test en Kaggle.
base_path = "/kaggle/input/dataset"

# Nombres de los splits que queremos procesar
splits = ["train", "dev", "test"]

def process_tokenized_folder(folder_path, output_csv):
    # Listamos únicamente los archivos que terminan en .tokenized en la carpeta.
    tokenized_files = [f for f in os.listdir(folder_path) if f.endswith('.tokens')]
    
    # Abrimos el archivo CSV de salida (uno por cada split), escribimos cabecera.
    with open(output_csv, "w", encoding="utf-8", newline='') as out_csv:
        writer = csv.writer(out_csv, delimiter='\t')
        writer.writerow(["id", "text", "label"])  # Escribimos la cabecera del CSV
        
        # Recorremos cada archivo .tokenized de la carpeta correspondiente.
        for filename in tokenized_files:
            file_path = os.path.join(folder_path, filename)
            file_prefix = os.path.splitext(filename)[0]  # Para construir el id: nombreArchivo-numOracion
            
            # Abrimos el archivo .tokenized
            with open(file_path, 'r', encoding='utf-8') as file:
                sentence_tokens = []  # Guarda los tokens de la oración actual
                sentence_labels = []  # Guarda las labels de la oración actual
                sentence_count = 1    # Contador de oraciones para el ID
                
                for line in file:
                    line = line.strip()
                    # Si la línea está vacía, fin de la oración actual
                    if line == "":
                        if sentence_tokens:  # Si hay tokens a procesar (oración no vacía)
                            # Si todas las labels son "O", la frase es no relevante (0); si alguna no es "O", relevante (1)
                            label = 0 if all(lbl == "O" for lbl in sentence_labels) else 1
                            # ID único formado por "nombreArchivo-numOracion"
                            row_id = f"{file_prefix}-{sentence_count}"
                            # Concatenamos los tokens con espacios para formar la oración original
                            text = " ".join(sentence_tokens)
                            # Escribimos fila en el CSV
                            writer.writerow([row_id, text, label])
                            sentence_count += 1
                            sentence_tokens = []
                            sentence_labels = []
                        continue  # Saltamos a la siguiente línea (siguiente oración)
                    try:
                        token, tag = line.rsplit(" ", 1)  # Separamos token y etiqueta (de derecha a izquierda por si el token contiene espacios)
                    except ValueError:
                        continue  # Saltar líneas mal formateadas
                    sentence_tokens.append(token)
                    sentence_labels.append(tag)
                # Por si la última oración no termina con salto de línea
                if sentence_tokens:
                    label = 0 if all(lbl == "O" for lbl in sentence_labels) else 1
                    row_id = f"{file_prefix}-{sentence_count}"
                    text = " ".join(sentence_tokens)
                    writer.writerow([row_id, text, label])

# Ejecutamos el procesamiento para cada split (train, dev, test)
for split in splits:
    tokenized_folder = os.path.join(base_path, split, "tokenized")
    output_csv = f"{split}.csv"
    process_tokenized_folder(tokenized_folder, output_csv)

)


```
