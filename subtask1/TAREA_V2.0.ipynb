```python
from transformers import AutoTokenizer

# Cargamos el tokenizador asociado a BERT base
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

```

```python
# Recalcular longitud máxima real del dataset
lengths = [len(tokenizer(text)["input_ids"]) for text in datasets_loaded["train"]["text"]]
print(f"Tamaño medio: {sum(lengths)//len(lengths)}, máximo real: {max(lengths)}")

# Usar como límite algo seguro
MAX_LENGTH = min(512, max(lengths))
print("Usando longitud máxima:", MAX_LENGTH)

```

```python
def tokenize(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=512
    )

encoded_data = {}
for split in datasets_loaded:
    print(f"🔄 Tokenizando split: {split}")
    encoded_data[split] = datasets_loaded[split].map(tokenize, batched=True)
```

```python
# ============================================================
# BACK TRANSLATION 50-50 (SOLO TRAINING)
# ============================================================

!pip install transformers sentencepiece sacremoses -q

import torch
from transformers import MarianMTModel, MarianTokenizer
from datasets import Dataset

class BackTranslation:
    def __init__(self, device='cuda' if torch.cuda.is_available() else 'cpu'):
        self.device = device

    def load_model(self, model_name):
        tokenizer = MarianTokenizer.from_pretrained(model_name)
        model = MarianMTModel.from_pretrained(model_name).to(self.device)
        return tokenizer, model

    def translate(self, texts, model, tokenizer, batch_size=32):
        translated = []
        for i in range(0, len(texts), batch_size):
            batch = texts[i:i+batch_size]
            inputs = tokenizer(batch, return_tensors="pt", padding=True,
                              truncation=True, max_length=512).to(self.device)
            with torch.no_grad():
                trans = model.generate(**inputs, max_length=512)
            translated.extend([tokenizer.decode(t, skip_special_tokens=True) for t in trans])
        return translated

    def back_translate(self, texts, target_lang="es"):
        print(f"🔄 EN → {target_lang} → EN")
        forward_tokenizer, forward_model = self.load_model(f'Helsinki-NLP/opus-mt-en-{target_lang}')
        translated = self.translate(texts, forward_model, forward_tokenizer)
        back_tokenizer, back_model = self.load_model(f'Helsinki-NLP/opus-mt-{target_lang}-en')
        return self.translate(translated, back_model, back_tokenizer)

# Separar clases
train_texts = datasets_loaded['train']['text']
train_labels = datasets_loaded['train']['label']

majority = [t for t, l in zip(train_texts, train_labels) if l == 0]
minority = [t for t, l in zip(train_texts, train_labels) if l == 1]

print(f"📊 Original: Clase 0: {len(majority)}, Clase 1: {len(minority)}")

# Calcular cuántos necesitamos
n_needed = len(majority) - len(minority)
print(f"🎯 Necesitamos {n_needed} ejemplos más de clase 1")

# Back Translation
bt = BackTranslation()
augmented_es = bt.back_translate(minority[:n_needed//2], "es")
augmented_fr = bt.back_translate(minority[:n_needed//2 + n_needed%2], "fr")

# Combinar
all_texts = majority + minority + augmented_es + augmented_fr
all_labels = [0]*len(majority) + [1]*(len(minority) + len(augmented_es) + len(augmented_fr))

print(f"\n✅ Balanceado: Clase 0: {all_labels.count(0)}, Clase 1: {all_labels.count(1)}")

# Tokenizar
balanced_dataset = Dataset.from_dict({'text': all_texts, 'label': all_labels})
train_dataset = balanced_dataset.map(tokenize, batched=True)
train_dataset = train_dataset.rename_column("label", "labels").remove_columns(["text"])
train_dataset.set_format("torch")

# Dev y test sin tocar
dev_dataset = encoded_data["dev"].remove_columns(["id", "text"]).rename_column("label", "labels")
test_dataset = encoded_data["test"].remove_columns(["id", "text"]).rename_column("label", "labels")
dev_dataset.set_format("torch")
test_dataset.set_format("torch")

print(f"✅ Train: {len(train_dataset)}, Dev: {len(dev_dataset)}, Test: {len(test_dataset)}")
```

```python
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

# Preparar datasets (SIN SMOTE)
train_dataset = encoded_data["train"].remove_columns(["id", "text"])
test_dataset = encoded_data["test"].remove_columns(["id", "text"])
dev_dataset = encoded_data["dev"].remove_columns(["id", "text"])

train_dataset = train_dataset.rename_column("label", "labels")
test_dataset = test_dataset.rename_column("label", "labels")
dev_dataset = dev_dataset.rename_column("label", "labels")

# Establecer formato torch
train_dataset.set_format("torch")
test_dataset.set_format("torch")
dev_dataset.set_format("torch")

# Calcular pesos de clase para el conjunto de entrenamiento
train_labels = [example['labels'].item() for example in train_dataset]

print("Distribución original de clases (train):")
unique, counts = np.unique(train_labels, return_counts=True)
for label, count in zip(unique, counts):
    print(f"  Clase {label}: {count} ejemplos")

# Calcular pesos balanceados
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(train_labels),
    y=train_labels
)

print(f"\nPesos de clase calculados: {class_weights}")
print(f"  Clase 0 (no relevante): peso = {class_weights[0]:.4f}")
print(f"  Clase 1 (relevante): peso = {class_weights[1]:.4f}")
```

#FINE TUNING

```python
from transformers import AutoModelForSequenceClassification

NUM_LABELS = 2  # porque tus etiquetas son binarias

model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=NUM_LABELS
)
```

###Hiperparámetros


```python
from transformers import TrainingArguments

# Creamos los hiperparámetros básicos del entrenamiento
args = TrainingArguments(output_dir="./outputs")

# Mostrar todos los hiperparámetros actuales
args
```

```python
args.per_device_train_batch_size = 32
args.per_device_eval_batch_size = 32
args.evaluation_strategy="epoch"

```

###Métricas

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support


def compute_metrics(pred):

    y_true = pred.label_ids                 # son las labels reales
    y_pred = pred.predictions.argmax(-1)    # son las predicciones


    acc = accuracy_score(y_true, y_pred)

    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='macro')

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

###Trainer

```python
from transformers import Trainer
from torch import nn
import torch

# Convertir class_weights a tensor
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float)

# Crear Trainer personalizado con pesos de clase
class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # Usar CrossEntropyLoss con pesos de clase
        loss_fct = nn.CrossEntropyLoss(weight=class_weights_tensor.to(model.device))
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss

print("✅ WeightedTrainer creado con class weights")
```

```python
# Usar el WeightedTrainer en lugar del Trainer normal
trainer = WeightedTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=dev_dataset,  # Usar dev_dataset que ya preparamos
    args=args,
    compute_metrics=compute_metrics,
    tokenizer=tokenizer  # Añadir el tokenizer
)

print("✅ Trainer inicializado con class weights")
print(f"📊 Train samples: {len(train_dataset)}")
print(f"📊 Dev samples: {len(dev_dataset)}")
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"

trainer.train()
```

```python
trainer.evaluate()
```

###EVALUACIÓN

```python
def get_prediction(text):
    # prepara el texto, aplicamos la misma tokenización que la utilizada en el training
    inputs = tokenizer(text, padding="max_length", max_length=MAX_LENGTH, truncation= True, return_tensors="pt").to("cuda")

    # aplicamos el modelo
    pred = model(**inputs).logits

    # obtenemos la probabilidad para cada clase
    probs = pred.softmax(1)
    # devolvemos la mayor
    return probs.argmax().item()
```

```python
y_pred=[get_prediction(text) for text in datasets_loaded['test']['text']]
y_true = datasets_loaded['test']['label']
```

```python
from sklearn.metrics import classification_report

LABELS = ["0 (no relevante)", "1 (relevante)"]
print(classification_report(y_true, y_pred, target_names=LABELS))
```
